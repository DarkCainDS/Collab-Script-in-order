```python
interesting_words = [(palabra, fdist[palabra]) for palabra in set(text1) if len(palabra)>5 and fdist[palabra]>10]
dtypes = [('word', 'S10'), ('frequency', int)]
interesting_words = np.array(interesting_words, dtype=dtypes)
ordered_words = np.sort(interesting_words, order = 'frequency')
ordered_words[-20:] # top de palabras mas frecuentes
```

```python
interesting_words
```

```python
top_word = 20
x = np.arange(len(ordered_words[-top_word:]))
y = [freq[1] for freq in interesting_words[-top_word:]]
plt.figure(figsize = (10,5))
plt.plot(x,y)
plt.xticks(x, [str(freq[0]) for freq in interesting_words[-top_word:]], rotation = 'vertical')
plt.grid(True)
plt.show()
```
